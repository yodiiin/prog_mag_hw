### **Задача 1**
Найдите в тексте все смайлики вида :), :(, :-), :-(.

```python
regexp = re.compile(r':-?[()]?')
text = '))():)) :( :-) :-(' # наверно :-))) не должно ловиться
print('Вот ваши смайлики:', *re.findall(regexp, text))
```

### Задача 2
Некоторые любят выделять особо важные места в тексте с помощью нижнего подчёркивания, например: "...здесь _очень важно_ отметить..." (дело в том, что нижние подчеркивания в некоторых редакторах автоматически превращаются в курсив).

Ищем такие выделения, но не длиннее 5 слов.

```python
regexp = re.compile(r'\b_\w+.?(?:\s\w+.?){1,4}_\b')
text = '...здесь _очень супер мега гипер важно_ отметить... print _Hello, world_!'
print('Выделенные словосочетания:')
for i in re.findall(regexp, text):
  print(i)
```

### Задача 3
Найдите в тексте все куски, которые взяты в круглые скобки.

```python
regexp = re.compile(r'\([\w\d\s.,!?;:\'"\{}\\/|=+-<>\[\]_]+\)') # запихнула туда знаки пунктуации
text = 'akdsj (asd; 932 3 p9f, 9w) kdk lwioqd (kkkskd123...) kaodokw (123)'
print('Скобочки:', *re.findall(regexp, text))
```

### Задача 5
В файле 26.txt содержатся сырые выкачанные из интернета новостные тексты. Извлеките сами тексты из файла. BeautifulSoup пользоваться пока нельзя.

```python
with open('26.txt', 'r', encoding='utf-8') as f:
  file_txt = f.read()

regexp = re.compile(r'<text><o>(.*?)</o></text>', re.DOTALL)
print(*re.findall(regexp, file_txt))
```

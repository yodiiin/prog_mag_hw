#### Задание 1. 

Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
import glob

students_info = glob.glob('C:/Users/Дина/Downloads/Data/Students_info_*.csv')
students_marks = glob.glob('C:/Users/Дина/Downloads/Data/Students_marks_*.csv')
df_info, df_marks = pd.DataFrame(), pd.DataFrame() # создаю отдельные датафремы для каждого "типа" имеющихся у нас табличек

for file in students_info: # объединяем все таблицы с информацией по студентам в один датафрейм
    data = pd.read_csv(file)
    df_info = pd.concat([df_info, data])

for file in students_marks: # объединяем все таблицы с оценками  один датафрейм
    data = pd.read_csv(file)
    df_marks = pd.concat([df_marks, data])

students = pd.merge(df_info, df_marks) # объединяем два больших датафрейма в одну таблицу

students.sample(5)
```

#### Задание 14. 

Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def overall_pass_status(row):
    if 'F' in [row['Math_PassStatus'], row['Reading_PassStatus'], row['Writing_PassStatus']]:
        return 'F'
    else:
        return 'P'

students['OverAll_PassStatus'] = students.apply(overall_pass_status, axis=1)
print(f'Число студентов, сдавших все экзамены: {len(students[students["OverAll_PassStatus"] == "P"])} (урааа!)')
students.sample(15)
```

#### Задание 15. 

Переведем баллы в оценки

*Система перевода баллов в оценки*

        [90;100] = A
        [80;90) = B
        [70;80) = C
        [60;70) = D
        [50;60) = E
        [0;50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
    if 90 <= average_mark <= 100:
        return 'A'
    elif 80 <= average_mark < 90:
        return 'B'
    elif 70 <= average_mark < 80:
        return 'C'
    elif 60 <= average_mark < 70:
        return 'D'
    elif 50 <= average_mark < 60:
        return 'E'
    elif 0 <= average_mark < 50:
        return 'F (fail)'
```

```python
students['Grade'] = students[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
students
```

```python
students['Grade'].value_counts()
```

1. Какова доля семей, в которых минимальный возраст меньше 20 (семьи с детьми)?

```python
titanic['Family'] = titanic['Name'].apply(lambda s: s.split(';')[0])
titanic.reset_index(drop=True, inplace=True)
titanic.head()
```

```python
children = titanic[titanic.Age < 20]
children
```

```python
families_total = len(set(titanic['Family']))
families_with_children = len(children.groupby('Family')['Age'].min() < 20)
```

```python
print(f'Доля семей, в которых минимальный возраст меньше 20 лет: {round((families_with_children / families_total)* 100, 2)}%')
```

2. Какова доля выживших пассажиров из класса 3? А пассажиров из класса 1?

```python
print(f'Количество выживших пассажиров: {len(titanic[titanic.Survived == 1])}\n\tИз них пассажиров 3 класса: {len(titanic[(titanic.Survived == 1) & (titanic.Pclass == 3)])}\n\tПассажиров 1 класса: {len(titanic[(titanic.Survived == 1) & (titanic.Pclass == 1)])}\n')
print(f'Доля выживших пассажиров из класса 3: {round((len(titanic[(titanic.Survived == 1) & (titanic.Pclass == 3)]) / len(titanic[titanic.Survived == 1])) * 100, 2)}%')
print(f'Доля выживших пассажиров из класса 1: {round((len(titanic[(titanic.Survived == 1) & (titanic.Pclass == 1)]) / len(titanic[titanic.Survived == 1])) * 100, 2)}%')
```

4. Создайте столбец "IsChild", который равен 1, если возраст меньше 20, и 0 иначе. Для пропущенных значений поведение функции может быть произвольным.

```python
titanic['IsChild'] = titanic['Age'].apply(lambda x: 1 if x < 20 else 0)
titanic.reset_index(drop=True, inplace=True)
```

5. Какова доля выживших женщин из первого класса? А доля выживших мужчин из 3 класса?

```python
survived_women = titanic[(titanic.Sex == 'female') & (titanic.Survived == 1)]
survived_men = titanic[(titanic.Sex == 'male') & (titanic.Survived == 1)]
survived_women_class1 = titanic[(titanic.Sex == 'female') & (titanic.Pclass == 1) & (titanic.Survived == 1)]
survived_men_class3 = titanic[(titanic.Sex == 'male') & (titanic.Pclass == 3) & (titanic.Survived == 1)]
print(f'Доля выживших женщин 1-го класса: {round(len(survived_women_class1) / len(survived_women) * 100, 2)}%')
print(f'Доля выживших мужчин 3-го класса: {round(len(survived_men_class3) / len(survived_men) * 100, 2)}%')
```

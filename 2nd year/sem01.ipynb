```python
import random
```

```python
class MusicStore:
    def __init__(self, name, account):
        self.name = name
        self.account = account
        self.warehouse = [] # сюда будем складывать музыкальные инструменты
    
    def sell_instrument(self, instrument, customer):
        if instrument not in self.warehouse: 
            print(f'Инструмента {instrument.name} нет на складе.\n')
        else:
            if instrument.is_broken:
                print(f'Инструмент {instrument.name} сломан. Необходимо починить его перед продажей!\n')
                self.fix_instrument(instrument)
            
            if customer.wallet >= instrument.cost:
                    self.warehouse.remove(instrument)
                    customer.wallet -= instrument.cost
                    self.account += instrument.cost
                    print(f'Инструмент {instrument.name} успешно приобретен!\n')
            else:
                print('У посетителя недостаточно средств для покупки.\n')
    
    def buy_instrument(self, instrument):
        if self.account >= instrument.cost:
            self.warehouse.append(instrument)
            self.account -= instrument.cost
            print(f'Инструмент {instrument.name} куплен за {instrument.cost} и добавлен на склад.\n')
        else:
            print('Недостаточно средств для покупки инструмента.\n')
    
    def fix_instrument(self, instrument):
        # if instrument not in self.warehouse:
        #     print(f'Инструмента {instrument.name} нет на складе.\n')
        # else:
            if instrument.is_broken:
                fix_cost = int(instrument.cost * 0.2)
                if self.account >= fix_cost:
                    instrument.is_broken = False
                    self.account -= fix_cost
                    print(f'Инструмент {instrument.name} починен за {fix_cost}.\n')
                else:
                    print('Недостаточно средств для ремонта.\n')
            else:
                print(f'Инструмент {instrument.name} в хорошем состоянии, ремонт не требуется.\n')
    
    def show_status(self):
        print(f'Баланс магазина: {self.account}')
        if self.warehouse:
            print('Склад магазина:')
            for instrument in self.warehouse:
                # print(f'\tИнструмент: {instrument.name} — Цена: {instrument.cost} — Состояние: {status}')
                print(f'\t{instrument.describe()}')
        else:
            print('Склад пуст.')
        print('\n')
```

```python
class MusicalInstrument:
    def __init__(self, name, cost):
        self.name = name
        self.cost = cost
        self.is_broken = random.choice([True, False])
    
    # def describe(self):
    #     pass
```

```python
class Guitar(MusicalInstrument):
    def __init__(self, name, cost):
        super().__init__(name, cost)
        types = ['акустическая', 'электрическая', 'классическая']
        self.type = random.choice(types)
        self.strings = 6
        bridges = ['кость', 'графит', 'никлеь', 'латунь']
        self.bridge = random.choice(bridges)
    
    def describe(self):
        status = 'сломан' if self.is_broken else 'исправен'
        return f"Инструмент: {self.name} {self.type}\n\tКоличество струн: {self.strings}\n\tТип порожка: {self.bridge}\n\tСтоимость: {self.cost}\n\tСостояние: {status}\n" 
        # print(f'Инструмент: {self.name} {self.type}')
        # print(f'Количество струн: {self.strings}')
        # print(f'Тип порожка: {self.bridge}')
        # print(f'Стоимость: {self.cost}')
```

```python
class Customer:
    def __init__(self, name, wallet):
        self.name = name
        self.wallet = wallet
    
    def show_wallet(self):
        print(f'{self.name} имеет в кошельке {self.wallet}\n')
    
    def buy_instrument(self, store):
        print(f'{self.name} хочет приобрести инструмент.')
        if store.warehouse:
            instrument = random.choice(store.warehouse)
            store.sell_instrument(instrument, self)
        else:
            print('На складе нет инстурментов.\n')
```

```python
class Game:
    def __init__(self):
        self.store = MusicStore('Музторг', 15000)
        self.customers = [
            Customer('Посетитель 1', random.randint(1000, 10000)),
            Customer('Посетитель 2', random.randint(1000, 10000)),
            Customer('Посетитель 3', random.randint(1000, 10000)),
            Customer('Посетитель 4', random.randint(1000, 10000))
            ]

    def working_day(self):
        for hour in range(8):
            print(f' -------- Час {hour + 1}-й -------- ')
            action = random.choice(['visitor', 'store_owner'])

            if action == 'visitor':
                customer = random.choice(self.customers)
                print(f'В магазин пришел {customer.name}.\n')
                customer.show_wallet()
                customer.buy_instrument(self.store)
            
            elif action == 'store_owner':
                print('Владелец магазина производит действие...\n')
                owner_action = random.choice(['buy_guitar', 'buy_bass', 'buy_ukulele', 'buy_piano', 'buy_drums', 'nothing'])

                if owner_action == 'buy_guitar':
                    guitar = Guitar('Гитара', 3500)
                    self.store.buy_instrument(guitar)
                
                elif owner_action == 'buy_bass':
                    bass = Bass('Бас-гитара', 3000)
                    self.store.buy_instrument(bass)
                
                elif owner_action == 'buy_ukulele':
                    ukulele = Ukulele('Укулеле', 1000)
                    self.store.buy_instrument(ukulele)
                
                elif owner_action == 'buy_piano':
                    piano = Piano('Пианино', 4500)
                    self.store.buy_instrument(piano)
                
                elif owner_action == 'buy_drums':
                    drums = Drums('Барабаны', 5000)
                    self.store.buy_instrument(drums)
                else:
                    print('Владелец решает ничего не предпринимать.\n')
            
            self.store.show_status()
        print('Рабочий день завершился!')
        return
    
    def play(self):
        print(f'Привет! Добро пожаловать в симулятор музыкального магазина {self.store.name}.\n')

        while True:
            command = input("Введи команду 'Начать день' или 'Выйти из игры': ").strip().lower()
            if command == 'начать день':
                self.working_day()
                break
            elif command == 'выйти из игры':
                print('\nВыход из игры. До встречи!')
                break
            else:
                print('Введена неверная команда. Попробуй еще раз.\n')
```
